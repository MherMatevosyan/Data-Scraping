```python
# Reading a csv file and plotting a column

import matplotlib.pyplot as plt

file=pd.read_csv("AirPassengers.csv")
plt.plot(file["Passengers"])
plt.show()
```

```python
# Converting string type of "date" to datetime type to be able to plot the values according to dates.

import matplotlib
import datetime

k=[]
for i in range(0,len(file["Month"])):
    dt = datetime.datetime.strptime(file["Month"][i], "%Y-%m")
    k.append(dt)
    
dates = matplotlib.dates.date2num(k)
plt.plot_date(dates, file["Passengers"])
plt.show()
```
