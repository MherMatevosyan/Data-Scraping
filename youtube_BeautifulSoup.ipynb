```python
import requests as req
from BeautifulSoup import *
import json
```

```python
url ='https://www.youtube.com/feed/trending'
re=req.get(url)
trend=re.text
soup=BeautifulSoup(trend)
```

```python
from pprint import pprint

X=soup.findAll('ul', attrs={'class':"expanded-shelf-content-list has-multiple-items"})
trending=[]
for j in range(len(X)):
    x=X[j].findAll('h3')
    y=X[j].findAll('div', attrs={'class':"yt-lockup-byline "})
    z=X[j].findAll('ul',attrs={'class':"yt-lockup-meta-info"})
    for i in range(len(x)):
        t={}
        link=x[i].find('a').get('href')
        title=x[i].find('a').get('title')
        dur=str(x[i].find('span').text)
        user=y[i].find('a').text
        channel=y[i].find('a').get('href')
        views=z[i].findAll('li')[1].text
        t['link'] = "https://www.youtube.com"+str(link)
        t['video'] = title
        t['duration'] = dur[11:-1]
        t['user_name'] = user
        t['channel'] = "https://www.youtube.com"+str(channel)
        t['views']  = views
        trending.append(t)
pprint(trending)
with open('trending_videos.json', 'w') as output:
    json.dump(trending, output, sort_keys = True, indent = 4)
```

```python
import unicodecsv as csv

with open('trending_videos.csv', 'w') as csv_file:
    for i in trending:
            writer = csv.writer(csv_file)
            for key, value in i.items():
                writer.writerow([key, value])
            writer.writerow([])
```
